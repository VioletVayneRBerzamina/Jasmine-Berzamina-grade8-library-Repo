```python
import firebase_admin
from firebase_admin import credentials, db
# Load the private key
cred = credentials.Certificate("/content/firebase_key.json")
# Initialize the app with your database URL
firebase_admin.initialize_app(cred, {
 "databaseURL": "https://grade8-json-db-default-rtdb.firebaseio.com/"
})
print("Firebase connected successfully!")
```

```python
import json

with open("/content/books.json", "r") as file:
    data = json.load(file)

print("JSON file loaded")

ref = db.reference("books")

for book in data["books"]:
    ref.child(str(book["id"])).set(book)

print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()

print("All Books Data:\n")

for book_id, book in books_data.items():
    print(book)
```

```python
for book_id, book in books_data.items():
    print(book["title"])
```

```python
for book_id, book in books_data.items():
    print(book["title"], "-", book["details"]["available"], "available")
```

```python
total_books = 0

for book_id, book in books_data.items():
    total_books += book["details"]["copies"]

print("Total number of books in the library:", total_books)
```
